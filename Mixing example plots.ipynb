```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
no_death_mixing = np.load('./model/no_death_outs_1/MixingAnalysisOnlypipe_out_Thu_Mar__1_11_07_26_2018/mixing_analysis.npy')
full_death_mixing = np.load('./model/full_death_outs_1/MixingAnalysisOnlypipe_out_Sun_Feb_18_19_17_09_2018/mixing_analysis.npy')
```

```python
no_death_mixing.shape
```

```python
no_death_mixing = no_death_mixing[np.argsort(no_death_mixing[:, 0, 0])]
```

```python
full_death_mixing = full_death_mixing[np.argsort(full_death_mixing[:, 0, 0])]
```

```python
distance_to_tumor_COM, distance_to_cluster_COM, p_sgas, n_sgas, p_pgas, n_pgas =  no_death_mixing[0]
distance_to_COM = distance_to_tumor_COM[0]
plt.scatter(np.argsort(distance_to_cluster_COM), p_pgas, label='No turnover (distance: {:3g})'.format(distance_to_COM))


distance_to_tumor_COM, distance_to_cluster_COM, p_sgas, n_sgas, p_pgas, n_pgas =  full_death_mixing[93]
distance_to_COM = distance_to_tumor_COM[0]
plt.scatter(np.argsort(distance_to_cluster_COM), p_pgas, c='r', label='Turnover (distance: {:3g})'.format(distance_to_COM))
plt.xlabel('Ranked Distance to COM of cluster')
plt.ylabel('% of private GAS')

plt.legend(loc=(0.1, 1))
```

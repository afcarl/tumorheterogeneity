```python
import json
```

```python
from glob import glob
```

```python
def get_folder_skeleton(death, surface_death=False):
    return '../MixingData/u002/1_{}_{}_outs_*/Jan18_MixingAnalysis*/mixing_analysis*.json'.format(int(surface_death), death.replace('.', ''))
```

```python
from itertools import product
```

```python
glob(get_folder_skeleton('0.2', False))
```

```python
death_rates = ['0', '0.05', '0.1','0.2', '0.65']
surface_turnover = [True, False]

data = {}
for params in product(death_rates, surface_turnover):
    try:
        data[params] = json.load(open(glob(get_folder_skeleton(*params))[0], 'r'))
    except IndexError as e:
        print(e)
```

```python
def get_cmap(n, name='Greys_r'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
from collections import defaultdict
from scipy.interpolate import interp1d
```

```python
cmap = get_cmap(10, name=u'ocean_r')

def plot_data(ax, data, title, cmap):
    datas_ = defaultdict(lambda: [])
    for freq_data in data:
        for i, size_ in enumerate(freq_data['sample_sizes']): 
            datas_[size_].append((freq_data['frequency'], freq_data['special_proportion'][i]))
            ax.errorbar(x=np.log10([freq_data['frequency']]),\
                        y=[1-freq_data['special_proportion'][i]],\
                        yerr=np.array([freq_data['std_special_proportion'][i]])/np.sqrt(100),\
#                         label='{}'.format(size_),
                        c=cmap(i+1),
                        marker='x',
                        alpha=0.3)
    
    for i, (size_, information) in enumerate(sorted(datas_.items(), key=lambda x:x[0])):
        x, y= zip(*information)
        x, y = np.array(x), 1-np.array(y)
        f = interp1d(x, y)
        x_range = np.linspace(np.min(x), np.max(x))
        ax.plot(np.log10(x_range),f(x_range), c=cmap(i+1), label='{}'.format(size_))
    ax.set_xlabel("$log_{10}(frequency)$")
    ax.set_ylabel('Mixing Ratio')
    ax.set_title(title)
    return ax
```

```python
f = plt.figure(figsize=(15, 3))    
ax = f.add_subplot(141)
plot_data(ax, data[('0.2', False)], 'Turnover, d=0.2', cmap)
# ax.legend()
```

```python

f = plt.figure(figsize=(15, 7))    
ax = f.add_subplot(241)
plot_data(ax, data[('0', False)], 'No Turnover', cmap)

ax.set_ylim([0, 1])

ax = f.add_subplot(242)
plot_data(ax, data[('0.05', True)], 'Surface Turnover, d=0.05', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(243)
plot_data(ax, data[('0.1', True)], 'Surface Turnover, d=0.1', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(244)
plot_data(ax, data[('0.2', True)], 'Surface Turnover, d=0.2', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(245)
plot_data(ax, data[('0.65', True)], 'Surface Turnover, d=0.65', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(246) 
plot_data(ax, data[('0.05', False)], 'Turnover, d=0.05', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(247)
plot_data(ax, data[('0.1', False)], 'Turnover, d=0.1', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(248)
plot_data(ax, data[('0.2', False)], 'Turnover, d=0.2', cmap)
ax.set_ylim([0, 1])

ax.legend([2, 6, 27, 50, 100, 1000, 10000, 20000], loc=(0.7, 0.2))
f.tight_layout()
f.savefig('mixing_analysis.pdf', dpi=150)
```

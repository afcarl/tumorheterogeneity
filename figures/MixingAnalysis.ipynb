```python
def get_cmap(n, name='Greys_r'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
from collections import defaultdict
from scipy.interpolate import interp1d
```

```python
def exponential_mean_function(x_i, y_i,k=1):
    """
        @params:
            x_i : all x_i points available
            y_i : the y_i values to average over
        @returns
            a function that can be evaluated for the mean at any point x
        
    """
    def exponential_mean_at(x):
        """
            @params:
                x: a np.array or float where the average value needs to be returned
            @returns
                <y(x)> : average value at position x
        """
        w = np.exp(-k*(x-x_i)**2)
        return np.dot(w, y_i) / np.sum(w)
    
    return exponential_mean_at
```

```python
cmap = get_cmap(10, name=u'ocean_r')

def plot_data(ax, data, title, cmap, k=0.1):
    datas_ = defaultdict(lambda: [])
    for freq_data in data:
        for i, size_ in enumerate(freq_data['sample_sizes']): 
            datas_[size_].append((freq_data['frequency'], freq_data['special_proportion'][i]))
            ax.errorbar(x=np.log10([freq_data['frequency']]),\
                        y=[1-freq_data['special_proportion'][i]],\
                        yerr=np.array([freq_data['std_special_proportion'][i]])/np.sqrt(100),\
#                         label='{}'.format(size_),
                        c=cmap(i+1),
                        marker='x',
                        alpha=0.3)
    
    for i, (size_, information) in enumerate(sorted(datas_.items(), key=lambda x:x[0])):
        x, y= zip(*information)
        x, y = np.array(x), 1-np.array(y)
#         print(x.shape)
        f = np.vectorize(exponential_mean_function(np.log10(x), y, k=k))
        x_range = np.linspace(np.min(x), np.max(x))
        ax.plot(np.log10(x_range),f(np.log10(x_range)), c=cmap(i+1), label='{}'.format(size_))
    ax.set_xlabel("$log_{10}(frequency)$")
    ax.set_ylabel('Mixing Ratio')
    ax.set_title(title)
    ax.set_ylim([0, 1])
    ax.set_xlim([-4.0, -1.0])
    return ax
```

```python
f = plt.figure(figsize=(15, 3))    
ax = f.add_subplot(141)
plot_data(ax, data[('0.2', False)], 'Turnover, d=0.2', cmap, k=12)
# ax.legend()
```

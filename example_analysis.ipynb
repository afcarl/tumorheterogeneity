```python
from itertools import cycle

plt.figure(figsize=(15,15))
radii = np.unique(df['radius'].values)
colors = cycle( cm.rainbow( np.linspace( 0, 1, len(radii) ) ) )

plt.subplot(2,2,1)

plt.title('Unique Drivers vs Distance')
for i in radii:
    subsample = df[df['radius'] == i]
    plt.scatter( subsample['distance_to_COM'], subsample['unique_driver_propotion'], color=next(colors), label='Radius='+str(i))
    plt.xlabel('Distance to Center of Mass')
    plt.ylabel('Proportion of Drivers that are Unique')

plt.subplot(2,2,2)

plt.title('Driver Enrichment vs Distance')
for i in radii:
    subsample = df[df['radius'] == i]
    plt.scatter( subsample['distance_to_COM'], subsample['driver_enrichment'], color=next(colors), label='Radius='+str(i))
    plt.xlabel('Distance to Center of Mass')
    plt.ylabel('Proportion of total SNPs that are driver')
    
plt.subplot(2,2,3)
plt.title('Unique Haplotypes vs Distance')
for i in radii:
    subsample = df[df['radius'] == i]
    plt.scatter( subsample['distance_to_COM'], subsample['unique_combos'], color=next(colors), label='Radius='+str(i))
    plt.xlabel('Distance to Center of Mass')
    plt.ylabel('Number of Unique Haplotypes')

plt.subplot(2,2,4)

plt.title('Proportion of SNPs that are Driver')
for i in radii:
    subsample = df[df['radius'] == i]
    plt.scatter( subsample['distance_to_COM'], subsample['proportion_driver'], color=next(colors), label='Radius='+str(i))
    plt.xlabel('Distance to Center of Mass')
    plt.ylabel('Proportion of Drivers')

plt.suptitle('Statistics about Driver SNPs in Samples', fontsize=15)
plt.legend(ncol=5, bbox_to_anchor=(0.65,2.40))
plt.show()
```

# Mixing analysis

```python
import json
```

```python
from glob import glob
```

```python
def get_folder_skeleton(death, surface_death=False):
    return './1_{}_{}_outs_*/Jan18_MixingAnalysis*/mixing_analysis*.json'.format(int(surface_death), death.replace('.', ''))
```

```python
from itertools import product
```

```python
death_rates = ['0', '0.05', '0.1','0.2', '0.65']
surface_turnover = [True, False]

data = {}
for params in product(death_rates, surface_turnover):
    try:
        data[params] = json.load(open(glob(get_folder_skeleton(*params))[0], 'r'))
    except IndexError as e:
        print(e)
```

```python
def get_cmap(n, name='Greys_r'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
cmap = get_cmap(10, name=u'ocean_r')

def plot_data(ax, data, title, cmap):
    for freq_data in data:
        for i, size_ in enumerate(freq_data['sample_sizes']): 
            ax.errorbar(x=np.log([freq_data['frequency']]),\
                        y=[freq_data['special_proportion'][i]],\
                        yerr=np.array([freq_data['std_special_proportion'][i]])/np.sqrt(100),\
                        label='{}'.format(size_), c=cmap(i+1), marker='x')
    ax.set_xlabel('Log_10(Frequency)')
    ax.set_ylabel('Special Proportion')
    ax.set_title(title)
```

```python

f = plt.figure(figsize=(15, 3))    
ax = f.add_subplot(141)
plot_data(ax, data[('0.05', False)], 'all death=0.05', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(142)
plot_data(ax, data[('0.1', False)], 'all death=0.1', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(143)
plot_data(ax, data[('0.2', False)], 'all death=0.2', cmap)
ax.set_ylim([0, 1])

ax = f.add_subplot(144)
# plot_data(ax, data[('0.65', False)], 'death=0.1', cmap)
ax.set_ylim([0, 1])

ax.legend([2, 6, 27, 50, 100, 1000, 10000, 20000], loc=(1, 0))
f.tight_layout()
```

```python

f = plt.figure(figsize=(15, 3))    
ax = f.add_subplot(141)
plot_data(ax, data[('0.05', True)], 'surface death=0.05', cmap)
ax.set_ylim([0, 1])
ax = f.add_subplot(142) 
plot_data(ax, data[('0.1', True)], 'surface death=0.1', cmap)
ax.set_ylim([0, 1])
ax = f.add_subplot(143)
plot_data(ax, data[('0.2', True)], 'surface death=0.2', cmap)
ax.set_ylim([0, 1])
ax = f.add_subplot(144)
plot_data(ax, data[('0.65', True)], 'surface death=0.65', cmap)
ax.set_ylim([0, 1])
ax.legend([2, 6, 27, 50, 100, 1000, 10000, 20000], loc=(1, 0))
f.tight_layout()
```
